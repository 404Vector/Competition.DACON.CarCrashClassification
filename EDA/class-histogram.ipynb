# Class Histogram Analysis

## Import Modules

```python
import os
import sys
sys.path.append('../')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
from Module import Result
```

## Load Data

```python
path_data = '../data/'
path_csv = os.path.join(path_data, 'train.csv')
df = pd.read_csv(path_csv)
results = [Result(label) for label in df['label']]
df['crash'] = [r.crash for r in results]
df['ego_involve'] = [r.ego_involve for r in results]
df['weather'] = [r.weather for r in results]
df['timing'] = [r.timing for r in results]
df.head()
```

```python
df.describe()
```

## Show Histogram

### Label Histogram

```python
fig, ax = plt.subplots(figsize=(12, 7))
plot = sbn.countplot(data=df, x='label', ax=ax)
for container in plot.containers:
    plot.bar_label(container)
plt.show()
```

### Crash Histogram

```python
fig, ax = plt.subplots(figsize=(12, 7))
plot = sbn.countplot(data=df, x='crash', ax=ax)
for container in plot.containers:
    plot.bar_label(container)
plt.show()
```

### Ego Involve Histogram

```python
fig, ax = plt.subplots(figsize=(12, 7))
plot = sbn.countplot(data=df, x='ego_involve', ax=ax)
for container in plot.containers:
    plot.bar_label(container)
plt.show()
```

### weather Histogram

```python
fig, ax = plt.subplots(figsize=(12, 7))
plot = sbn.countplot(data=df, x='weather', ax=ax)
for container in plot.containers:
    plot.bar_label(container)
plt.show()
```

### Timing Histogram

```python
fig, ax = plt.subplots(figsize=(12, 7))
plot = sbn.countplot(data=df, x='timing', ax=ax)
for container in plot.containers:
    plot.bar_label(container)
plt.show()
```
